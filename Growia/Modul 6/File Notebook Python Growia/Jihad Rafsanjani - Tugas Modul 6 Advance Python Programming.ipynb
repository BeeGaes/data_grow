```python
#1. visualisasi untuk menunjukkan gross profit setiap quarternya.

df['quarter'] = df['date'].dt.to_period('Q')  # Menghitung quarter
df['quarter'] = df['quarter'].dt.strftime('%q')  # Mengambil angka quarter
df['year'] = df['date'].dt.year  # Mengambil tahun

# Agregasi gross profit berdasarkan quarter dan tahun
df_agg = df.groupby(['year', 'quarter'])['gross_profit'].sum(numeric_only=True).reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_agg['year'].astype(str) + '-' +df_agg['quarter'], df_agg['gross_profit'],
         marker='o', linestyle='-')
plt.title('Gross Profit per Quarter')
plt.xlabel('Quarter and Year')
plt.ylabel('Gross Profit (millions)')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_agg['year'].astype(str) + '-' + df_agg['quarter'], df_agg['gross_profit'],
         marker='o', linestyle='-')
plt.title('Gross Profit per Quarter')
plt.xlabel('Quarter and Year')
plt.ylabel('Gross Profit (millions)')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Menambahkan penanda pandemi COVID-19 pada pertengahan 2020
plt.axvspan('2020-2', '2020-3', color='red', alpha=0.3, label='COVID-19 Pandemic')

plt.legend()
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Membuat subplot
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot Total Revenue vs Cost of Revenue
axs[0].scatter(df['total_revenue'], df['cost_of_revenue'], color='blue')
axs[0].set_title('Total Revenue vs Cost of Revenue')
axs[0].set_xlabel('Total Revenue')
axs[0].set_ylabel('Cost of Revenue')
axs[0].grid(True, linestyle='--', alpha=0.7)

# Scatter plot Premium Revenue vs Premium Cost of Revenue
axs[1].scatter(df['premium_revenue'], df['premium_cost_revenue'], color='green')
axs[1].set_title('Premium Revenue vs Premium Cost of Revenue')
axs[1].set_xlabel('Premium Revenue')
axs[1].set_ylabel('Premium Cost of Revenue')
axs[1].grid(True, linestyle='--', alpha=0.7)

# Scatter plot Ad Revenue vs Ad Cost of Revenue
axs[2].scatter(df['ad_revenue'], df['ad_cost_of_revenue'], color='red')
axs[2].set_title('Ad Revenue vs Ad Cost of Revenue')
axs[2].set_xlabel('Ad Revenue')
axs[2].set_ylabel('Ad Cost of Revenue')
axs[2].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
